# Importing Data from a CSV file

```python
#import pandas
import pandas as pd
```

```python
#Import the CSV in a dataframe
cities=pd.read_csv("cities.csv")
cities
```

```python
#Add a column to the dataframe
cities['Pincode']=''
```

```python
#Perform describe and info on dataframe
cities.describe()
```

```python
#Provides all the information about dataframe including number of rows and columns
cities.info
```

```python
#Export and save this data as cities1.csv
cities.to_csv("cities1.csv")
```
